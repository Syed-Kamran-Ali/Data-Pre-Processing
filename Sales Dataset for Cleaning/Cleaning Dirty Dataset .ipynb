```python
import pandas as pd 
```

```python
dataset = pd.read_excel("Ditry Data Sample.xlsx")
```

```python
dataset = dataset[:-1]
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
dataset.info()
```

# Filling Empty Cells

```python
dataset.fillna("-") 
```

# Seperating Consumer,Corporate and Home Data

```python
consumer_index = dataset[dataset["Consumer Total"].notnull()].index
```

```python
corporate_index = dataset[dataset["Corporate Total"].notnull()].index
```

```python
home_index = dataset[dataset["Home Office Total"].notnull()].index
```

# Creating New Columns For Each 

```python
dataset["Segment"] = "a"
dataset["Order_ID"] = "b"
dataset["Ship_Mode"] = "c"
dataset["Sales"] = "d"
```

# Filling Specific Values in specific loactions

```python

#Filling Consumer Segment,Order ID,Sales Column
for i in consumer_index:
    dataset.loc[i,"Segment"]= "Consumer"
    dataset.loc[i,"Order_ID"] = dataset.loc[i][0]
    dataset.loc[i,"Sales"] = dataset.loc[i][5]
    
# Filling Ship_mode Coloum
for i in consumer_index:
    for j in range(1,5):
        if dataset.loc[i][j] >0:
            dataset.loc[i,"Ship_Mode"] = dataset.loc[0][j]   

      
```

```python
dataset.head(10)
```

```python

#Filling Corporate Segment,Order ID and Sales Column
for i in corporate_index:
    dataset.loc[i,"Segment"]= "Corporate"
    dataset.loc[i,"Order_ID"] = dataset.loc[i][0]
    dataset.loc[i,"Sales"] = dataset.loc[i][10]
    
# Filling Ship_mode Coloum
for i in corporate_index:
    for j in range(6,10):
        if dataset.loc[i][j] >0:
            dataset.loc[i,"Ship_Mode"] = dataset.loc[0][j]   

      
```

```python
dataset.head(10)
```

```python

#Filling Home Office Segment,Order ID and Sales Column
for i in home_index:
    dataset.loc[i,"Segment"]= "Home Office"
    dataset.loc[i,"Order_ID"] = dataset.loc[i][0]
    dataset.loc[i,"Sales"] = dataset.loc[i][15]
    
# Filling Ship_mode Coloum
for i in home_index:
    for j in range(11,15):
        if dataset.loc[i][j] >0:
            dataset.loc[i,"Ship_Mode"] = dataset.loc[0][j]   

      
```

```python
dataset[dataset["Segment"] == "Home Office"]
```

# Filtering and Grouping data 

```python
dataset_new = dataset[["Segment","Order_ID","Ship_Mode","Sales"]][2:]
```

```python
dataset_new.reset_index(drop = True,inplace=True)
```

```python
dataset_new = dataset_new.groupby(["Segment", "Ship_Mode"])
```

```python
dataset_new.groups.keys()
```

```python
a = [dataset_new.get_group(i) for i in dataset_new.groups.keys()]
```

```python
dataset_new = pd.concat(a,ignore_index=True)
```

```python
dataset_new
```

# Pivot Table to understand dynamics of data

```python
dataset_new.pivot_table(index = ["Segment", "Ship_Mode"], values = "Sales", aggfunc = "sum")
```

```python
dataset_new.to_excel("Cleaned Data.xlsx",index = False)
```
